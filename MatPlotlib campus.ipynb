```python
# from pandas dataframes

batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# give labels and title
plt.figure(figsize=(12,6))
plt.plot(batsman['index'],batsman['V Kohli'], label='Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit')
plt.title("Rohis Vs V Kohli Year wise Runs")
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
```

```python
# change color and line style
plt.figure(figsize=(12,6))
plt.plot(batsman['index'],batsman['V Kohli'],color='blue' ,label='Kohli',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',label='Rohit', linestyle='dotted')
plt.title("Rohis Vs V Kohli Year wise Runs")
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
```

```python
# add linewidth and marker point style
plt.figure(figsize=(12,6))
plt.plot(batsman['index'],batsman['V Kohli'],color='blue' ,label='Kohli',linestyle='dashed',marker='*',linewidth=2,markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',label='Rohit', linestyle='dotted',marker='h',linewidth=2,markersize=10)
plt.title("Rohis Vs V Kohli Year wise Runs")
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
# without using marker, color, line style, just pass in location 3 parameter 'Dr--' dymond marker,r for red and dashed line
plt.figure(figsize=(12,6))
plt.plot(batsman['index'],batsman['V Kohli'],'Dr--' ,label='Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],'*b-.',label='Rohit')
plt.title("Rohis Vs V Kohli Year wise Runs" , color='k',size=25,pad=15) # pad for title distance from graph
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
```

## Subplots


```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['avg'],y=df['strike_rate'])
plt.title('Avg and Strite analysis on Top 50 Batsman',size=15, pad=15)
plt.ylabel('Strike Rate')
plt.xlabel('Average')
plt.show()
```

```python
# subplot method
fig,ax = plt.subplots(figsize=(10,6)) # gives two object one is fig and 2nd is axis
ax.scatter(df['avg'],df['strike_rate'],c='r',marker='+')
ax.set_title('Avg and Strite analysis on Top 50 Batsman',c='g')
ax.set_xlabel('average')
ax.set_ylabel('strike Rate')
fig.show()
```

```python
fig,ax = plt.subplots(2,1,figsize=(12,6),sharex=True)  # gives 1 figure and 2 axes
ax[0].scatter(df['avg'],df['strike_rate'],c='g')
ax[0].set_title('Avg Vs Strike_rate',size=15)
ax[0].set_ylabel('Strike_rate')


ax[1].scatter(df['avg'],df['runs'],c='r')
ax[1].set_title('Avg Vs Runs',size=15)
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Average')

plt.show()
```

```python
fig,ax = plt.subplots(2,2,figsize=(12,6))  # gives 1 figure and 4 axes

ax[0,0].scatter(df['avg'],df['strike_rate'],c='g')
ax[0,1].scatter(df['avg'],df['runs'],c='r')
ax[1,0].hist(df['runs'])
ax[1,1].hist(df['avg'])

ax[0,0].set_title('Avg Vs Strike_rate',size=15)
ax[0,1].set_title('AVg vs Runs', size=15)    
ax[1,0].set_title('Runs Frequency',size=15)
ax[1,1].set_title('avg Frequency',size=15)                  
```

```python
## second method 

fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(df['avg'],df['strike_rate'])

ax2 = fig.add_subplot(2,2,2)
ax2.hist(df['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(df['avg'])
```

## 3D surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(14,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
plt.show()
```

## contour Plots

```python
fig = plt.figure(figsize=(14,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
z=np.tan(xx)+np.cos(yy)
fig = plt.figure(figsize=(14,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
plt.show()
```
